```python
import numpy as np
from matplotlib import pyplot as plt

plt.style.use('seaborn-whitegrid')
```

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0,10,1000)
ax.plot(x, np.sin(x));
```

```python
plt.plot(x, np.sin(x));
```

```python
plt.plot(x, np.sin(x));
plt.plot(x, np.cos(x));
```

```python
plt.plot(x, x**2, color='green', linestyle='-');
plt.plot(x, x*np.log(x+1), color='grey', linestyle='--');
plt.plot(x, 1.2*x, color='grey', linestyle='-.');
plt.plot(x, x, color='red', linestyle=':');
```

```python
plt.plot(x, np.sin(x), 'bo');
plt.plot(x, np.cos(x), 'r+');
```

```python
plt.plot(x, np.sin(x))
plt.xlim(-1,11);
plt.ylim(-1.5,1.5);
```

```python
plt.plot(x, np.sin(x))
plt.axis([-1,11,-1.5,1.5]);
```

```python
plt.plot(x, np.sin(x))
plt.axis('tight');
```

```python
plt.plot(x, np.sin(x))
plt.axis(False);
```

```python
plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.plot(x, np.zeros(1000,), color='red');
```

```python
plt.plot(x, np.sin(x), color='green', linestyle='-', label='sin(x)')
plt.plot(x, np.cos(x), color='gray', linestyle=':', label='cos(x)')
plt.axis('equal')
plt.legend();
```

```python
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(
    xlabel='x',
    ylabel='sin(x)',
    xlim=(5,10),
    ylim=(0,1),
    title='A Distorted Plot'
);
```
